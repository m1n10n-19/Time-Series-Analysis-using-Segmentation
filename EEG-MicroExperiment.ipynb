```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
music_data = pd.read_csv('tiny_eeg_self_experiment_music.csv')
reading_data = pd.read_csv('tiny_eeg_self_experiment_reading.csv')
```

```python
final_music = music_data.iloc[5857:,:]
final_reading = reading_data[reading_data['Channel2']>=20]
final_reading = final_reading[final_reading['Channel2']<=360]
```

```python
final_music.head()
```

```python
final_reading.head()
```

```python
df_music = final_music.drop(columns = ['IndexId', 'Ref1', 'Ref2', 'Ref3', 'TS1', 'TS2'])
df_read = final_reading.drop(columns = ['IndexId', 'Ref1', 'Ref2', 'Ref3', 'TS1', 'TS2'])
```

```python
df_read['time'] = df_read.reset_index().index
df_music['time'] = df_music.reset_index().index
```

```python
x = 0

for i in range(20):
    xgraph = df_read.time[x:(x+100)]
    ygraph = df_read.Channel2[x:(x+100)]

    plt.figure(figsize=(15, 4))
    plt.axis('off')
    plt.plot(xgraph, ygraph)
    
    plt.savefig(f'MicroExperiment_Figures/reading_{x}_graph.png')
    
    x = x + 100
```
