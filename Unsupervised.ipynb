```python
class tree(object):
    def __init__(self, data_point):
        self.data_point = data_point
        self.children = []
        self.end = 0
        self.freq = 0
```

```python
def subarray(string):
  f = open("substring_1.txt", "w+")
  c = 0
  for i in range(len(string)):
    for j in range(2,min(15,len(string)-i+1)):               
      try:
        f.writelines(str(string[i:i+j]).replace('0','*')+'\n')
        c += 1
      except:
        pass
  f.close()
```

```python
def entropy(arr):
    s = 0
    for i in arr:
      if i != 0:
        s += -i / sum(arr)*math.log2(i/sum(arr))
    return s
```

```python
def add(root, word: str, variation, frequency):
    node = root
    for data_point in word:
        found_in_child = False
        # Search for the character in the children of the present `node`
        for child in node.children:
            if child.data_point == data_point:
              node = child
              found_in_child = True
              break
        # We did not find it so add a new chlid
        if not found_in_child:
          new_node = tree(data_point)
          node.children.append(new_node)
          # And then point node to the new child
          node = new_node
    node.counter=variation
    node.freq=frequency
```

```python
def primitives(d):
  root = tree('*')
  for i in d.keys():
    add(root, i, d[i][0],d[i][1])
  return root
```

```python
def print_tree(root,s,final,inter,prev_c,prev_f,length):
  s+=root.data_point
  if root.counter>prev_c:
    final.append(s[1:])
    print(s[1:],root.counter,prev_c)
  if root.children==[]:
    s=''
    length=0
    return 0
  for child in root.children:
    print_tree(child,s,final,inter,root.counter,root.freq,length+1)
```
